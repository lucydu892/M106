## Aufgabe 11

```sql
CREATE VIEW PlaylistTracks AS
SELECT PL.PlaylistId, PL.TrackId, T.Name, IL.UnitPrice FROM PlaylistTrack PL 
JOIN Track T ON T.TrackId = PL.TrackId
JOIN InvoiceLine IL ON T.TrackId = IL.TrackId
GROUP BY PL.PlaylistId
ORDER BY ACS
```

```sql
SELECT * FROM PlaylistTracks GROUP BY PlaylistTrack.PlaylistId
ORDER BY ACS;
```
